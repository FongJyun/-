```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import insightface
from insightface.app import FaceAnalysis
from insightface.data import get_image as ins_get_image
import matplotlib.patches as patches

model_pack_name = 'buffalo_sc'
app = FaceAnalysis(name=model_pack_name, allowed_modules=['detection','recognition'])
app.prepare(ctx_id=0, det_size=(256, 256))
```

```python
import inspect
inspect.getargspec(FaceAnalysis)
```

```python
import os
[i for i in os.listdir('.') if i.endswith('png') | i.endswith('jpg')]
```

```python

img = cv2.imread('uHnRREJ.jpg')
img = cv2.resize(img, (512,512))
plt.imshow(img[...,::-1])
```

```python
img_emb_results = app.get(np.asarray(img))
img_emb = img_emb_results[0]
img_emb['embedding'].shape

who_ever = img_emb
```

```python
fig, ax = plt.subplots()

ax.imshow(img[...,::-1])

def tobox(img_emb_bbox):
  img_emb_bbox = img_emb_bbox.astype(int)
  origin = (img_emb_bbox[0], img_emb_bbox[1])
  w = img_emb_bbox[2] - img_emb_bbox[0]
  h = img_emb_bbox[3] - img_emb_bbox[1]
  return origin, w, h

o,w,h = tobox(who_ever.bbox)
# Create a Rectangle patch
rect = patches.Rectangle(o, w, h, linewidth=1, edgecolor='r', facecolor='none')
# Add the patch to the Axes
ax.add_patch(rect)


plt.show()
```

```python
plt.figure()

plt.subplot(121)
chou_1 = cv2.imread('chou_1.png')
chou_1 = cv2.resize(chou_1, (512,512))
plt.imshow(chou_1[...,::-1])

plt.subplot(122)
chou_2 = cv2.imread('chou_2.png')
chou_2 = cv2.resize(chou_2, (512,512))
plt.imshow(chou_2[...,::-1])
plt.show()
```

```python
chou_lis = [chou_1, chou_2]
lis = []

def tobox(img_emb_bbox):
  img_emb_bbox = img_emb_bbox.astype(int)
  origin = (img_emb_bbox[0], img_emb_bbox[1])
  w = img_emb_bbox[2] - img_emb_bbox[0]
  h = img_emb_bbox[3] - img_emb_bbox[1]
  return origin, w, h

for i in chou_lis:

  img_emb_results = app.get(np.asarray(i))
  img_emb = img_emb_results[0]
  

  fig, ax = plt.subplots()

  ax.imshow(i[...,::-1])

  o,w,h = tobox(img_emb.bbox)
  # Create a Rectangle patch
  rect = patches.Rectangle(o, w, h, linewidth=1, edgecolor='r', facecolor='none')
  # Add the patch to the Axes
  ax.add_patch(rect)


  plt.show()

  lis.append(img_emb)
```

## bigger the best

```python
cos_sim = lambda x1,x2: np.dot(x1,x2) / (np.linalg.norm(x1)*np.linalg.norm(x2))
print(cos_sim(lis[0]['embedding'], lis[1]['embedding']) )
print(cos_sim(lis[0]['embedding'], who_ever['embedding']))
```
